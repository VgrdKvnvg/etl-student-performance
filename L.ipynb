```python
from pyspark.sql.functions import avg, count

spark.sql("DROP TABLE IF EXISTS gold_hours_vs_score")

df = spark.table("silver_clean")

gold_df = df.groupBy("Hours_Studied") \
    .agg(
        avg("Exam_Score").alias("avg_exam_score"),
        count("*").alias("n")
    ) \
    .filter("n >= 5") \
    .orderBy("Hours_Studied")

gold_df.write.mode("overwrite").saveAsTable("gold_hours_vs_score")
```
